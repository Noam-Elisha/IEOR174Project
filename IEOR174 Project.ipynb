```python
import random
import numpy as np
import csv
import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = datetime.datetime(2019,10,3)
def get_time(date, hour):
    date = date.split('/')
    temp = datetime.datetime(int(date[2]),int(date[0]),int(date[1]), int(hour[:2]))
    diff = temp-base
    days, seconds = diff.days, diff.seconds
    hours = days * 24 + seconds // 3600
    minutes = (seconds % 3600) // 60
    seconds = seconds % 60
    return hours   
    
```

```python
csv_input = pd.read_csv('GeneralData.csv')
temp = csv_input[['date', 'hour']]
csv_input['time'] = [get_time(date, hour) for date, hour in zip(temp['date'], temp['hour'])]
csv_input.to_csv('GeneralData1.csv', index=False)

csv_input = pd.read_csv('CitizenData.csv')
temp = csv_input[['date', 'hour']]
csv_input['time'] = [get_time(date, hour) for date, hour in zip(temp['date'], temp['hour'])]
csv_input.to_csv('CitizenData1.csv', index=False)

csv_input = pd.read_csv('NonCitizenData.csv')
temp = csv_input[['date', 'hour']]
csv_input['time'] = [get_time(date, hour) for date, hour in zip(temp['date'], temp['hour'])]
csv_input.to_csv('NonCitizenData1.csv', index=False)
```

```python
general_data = pd.read_csv('GeneralData1.csv')
citizen_data = pd.read_csv('CitizenData1.csv')
non_citizen_data = pd.read_csv('NonCitizenData1.csv')
```

```python
plt.plot(general_data['time'], general_data['Average Wait Time'], 'x')
plt.title("Average waiting time for a passenger through customs over time")
plt.show()
plt.plot(citizen_data['time'], citizen_data['Average Wait Time'], 'x')
plt.title("Average waiting time for a citizen through customs over time")
plt.show()
plt.plot(non_citizen_data['time'], non_citizen_data['Average Wait Time'], 'x')
plt.title("Average waiting time for a non-citizen through customs over time")
plt.show()
```
